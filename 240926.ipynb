```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)

plt.plot(x,y, x,z)
plt.xlabel('Time axis')
plt.ylabel('magnitude')
plt.title('Cose & Sine Waves')
plt.legend(['Y-value', 'Z-value'], loc=4)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'r-x')
plt.title('Cosine Wave')
plt.subplot(2,1,2)
plt.plot(x,z, 'g-o')
plt.title('Sine Wave')

plt.tight_layout()
plt.savefig('./drive/MyDrive/test0.jpg', dpi=300)
```
