```python
import numpy as np
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import pandas as pd
import time
from sklearn.model_selection import train_test_split
startTime = time.time()
data = pd.read_csv('weatherHistory.csv')
data.head()
```

```python
data.info()
data['Precip Type'] = data['Precip Type'].map( {'rain':0, 'snow':1})
```

```python
data['Precip Type'].fillna(value=0,inplace=True)
data['Precip Type'].astype(int)
```

```python
x = data[['Temperature (C)', 'Humidity']].loc[1:90000].values.reshape((90000, 2))
y = data['Precip Type'].loc[1:90000]
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2, shuffle=True)
clf=make_pipeline(StandardScaler(),
                  SGDClassifier(loss='log',max_iter=1000, tol = 1e-3))
clf.fit(train_x, train_y)
print(clf.score(train_x,train_y))
print(clf.score(test_x,test_y))
```

```python
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
data = pd.read_csv('weatherHistory.csv')
data['Precip Type'] = data['Precip Type'].map({'rain':0, 'snow':1})
data['Precip Type'].fillna(value=0, inplace=True)
x = data[ ['Temperature (C)', 'Humidity'] ]
y = data['Precip Type']
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, shuffle=True)
clf = make_pipeline(StandardScaler(), svm.SVC(kernel='rbf', C=1.0, gamma=1))
clf.fit(train_x, train_y)
print(clf.score(train_x, train_y))
print(clf.score(test_x, test_y))
y_pred = clf.predict(test_x)
print( confusion_matrix(test_y, y_pred) )
print( precision_score(test_y, y_pred) )
print( recall_score(test_y, y_pred, average='macro') )
print( accuracy_score(test_y, y_pred) )
print( f1_score(test_y, y_pred, average='weighted') )
```
