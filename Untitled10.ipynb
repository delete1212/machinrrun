```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

svr_data = pd.read_csv('gun_deaths.csv')

svr_data = svr_data.dropna().copy()
svr_data['intent']=svr_data['intent'].map({'Suicide':1,'Undetermined':2, 'Accidental':3, 'Homicide':4})

data = svr_data[['age']]
target = svr_data['intent']

train_data,test_data,train_target,test_target = train_test_split(data, target, test_size=0.2,
                                                                 shuffle=False)

from sklearn.preprocessing import StandardScaler
st=StandardScaler()
train_data = st.fit_transform(train_data)
test_data = st.transform(test_data)

sampleNum = 1000

train_data_sample = train_data[:sampleNum]
train_target_sample = train_target[:sampleNum]

from sklearn import svm
clf_svr_linear = svm.SVR(kernel='linear',C=1.0)
clf_svr_linear.fit(train_data_sample,train_target_sample)


r_linear = clf_svr_linear.predict(test_data[:5])
print(r_linear)
print(test_target[:5])
print(test_data[:5])

#나이 32
print('나이 32세일 때 뭘로 총기사건을 당했는지 예측')
print(clf_svr_linear.predict(st.transform([[32]])))


clf_svr_poly = svm.SVR(kernel='poly', C=1.0)
clf_svr_poly.fit(train_data_sample, train_target_sample)
r_poly = clf_svr_poly.predict(train_data_sample)
print(r_poly)

clf_svr_rbf = svm.SVR(kernel='rbf', C=1.0,gamma=1)
clf_svr_rbf.fit(train_data_sample,train_target_sample)
r_rbf = clf_svr_rbf.predict(test_data)
print(r_rbf)

print('svr_linear')
print(clf_svr_linear.predict(st.transform([[32]])))

print('svr_poly')
print(clf_svr_poly.predict(st.transform([[32]])))

print('svr_rbf')
print(clf_svr_rbf.predict(st.transform([[32]])))
```
