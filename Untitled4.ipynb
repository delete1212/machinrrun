```python
from sklearn.datasets import load_breast_cancer
bc_data = load_breast_cancer()

import pandas as pd
df_data=pd.DataFrame(bc_data.data)
df_labels = pd.DataFrame(bc_data.target)

print(df_labels.head())
print(bc_data.target_names)
```

```python
df_data.head()
```

```python
df_data.describe()
```

```python
def min_max_normalize(lst):
  normalized=[]
  for value in lst:
    normalized_num = (value-min(lst)) / (max(lst)-min(lst))
    normalized.append(normalized_num)

    return normalized
```

```python
for x in range(len(df_data.columns)):
  df_data[x] = min_max_normalize(df_data[x])

df_data.describe()
```

```python
from sklearn.model_selection import train_test_split

training_data, validation_data, training_labels, validation_labels \
= train_test_split(df_data, df_labels, test_size=0.2, random_state=100)

print(len(training_data))
print(len(validation_data))
print(len(training_labels))
print(len(validation_data))
```

```python
from sklearn.neighbors import KNeighborsClassifier
cf = KNeighborsClassifier(n_neighbors=3)
cf.fit(training_data, training_labels)
print(cf.score(validation_data, validation_labels))
```

```python
import matplotlib.pyplot as plt
k_list=range(1, 101)
accuracies = []

for k in k_list:
  cf = KNeighborsClassifier(n_neighbors= k)
  cf.fit(training_data, training_labels)
  accuracies.append(cf.score(validation_data, validation_labels))
plt.plot(k_list, accuracies)
plt.xlabel("k")
plt.ylabel("validation accuracy")
plt.show()
```

```python
for k in range(5):
  cf = KNeighborsClassifier(n_neighbors=k+1)
  cf.fit(training_data, training_labels)
  print(cf.score(validation_data, validation_labels))
```
