```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/titanic_train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(value=df['Age'].mean(), inplace=True)
df.isnull().sum()
```

```python
df.info
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 1, 'C': 2, 'Q': 3})
df['Embarked'].fillna(value=3, inplace=True)
```

```python
feature = df[['Sex', 'Age', 'Pclass', 'Embarked']]
target = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np

feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.2, random_state=42)
```

```python
ss = StandardScaler()
feature_train = ss.fit_transform(feature_train)
feature_test = ss.transform(feature_test)
clf=LogisticRegression()
clf.fit(feature_train, target_train)
```

```python
print(clf.score(feature_train, target_train))
print(clf.score(feature_test, target_test))
```

```python
Jack = np.array([0.0, 20.0, 2.0, 2.0])
Rose = np.array([1.0, 17.0, 1.0, 1.0])
ME = np.array([0.0, 47.0, 3.0, 3.0])
sample= np.array([Jack, Rose, ME])
sample = ss.transform(sample)
pred = clf.predict(sample)
print(pred)
print(clf.predict_proba(sample))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('/content/drive/MyDrive/weatherHistory.csv')
data.head()
```

```python
data['Precip Type'].fillna(value='rain', inplace=True)
data['Precip Type'] = data['Precip Type'].map({'rain': 0, 'snow': 1})
```

```python
feature = data[['Temperature (C)', 'Humidity', 'Wind Speed (km/h)']]
target = data['Precip Type']
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.2, shuffle=False)
```

```python
ss=StandardScaler()
feature_train = ss.fit_transform(feature_train)
feature_test = ss.transform(feature_test)
```

```python
clf = LogisticRegression()
clf.fit(feature_train, target_train)
print(clf.score(feature_train, target_train))
print(clf.score(feature_test, target_test))
```

```python
DU = np.array([32.0, 0.65, 2])
Pohang = np.array([15.1, 0.84, 2.4])
Seoul = np.array([24.1, 0.75, 2.5])
SoeulWinter = np.array([-6.0, 0.74, 2.9])
Daegu = np.array([0.0, 0.25, 2])
sample = np.array([DU, Seoul, Pohang, SoeulWinter, Daegu])
sample = ss.transform(sample)
pred = clf.predict(sample)
print(pred)
print(clf.predict_proba(sample))
```
