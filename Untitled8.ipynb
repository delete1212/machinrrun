```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

death = pd.read_csv('gun_deaths.csv')

gun_death = death[death['age'] >= 1]

gun_death['month'] = gun_death['month'].astype(int)

X = np.array(gun_death['month'])
y = np.array(gun_death['age'])
xx = X[:, np.newaxis]
yy = y[:, np.newaxis]

random_indices = np.random.choice(len(xx), size=100, replace=False)
xx_sampled = xx[random_indices]
yy_sampled = yy[random_indices]

model = LinearRegression().fit(xx_sampled, yy_sampled)

month_grid = np.linspace(X.min(), X.max(), 12)[:, np.newaxis]

predicted_age = model.predict(month_grid)

plt.scatter(xx_sampled, yy_sampled, color='black', label='Sampled Data')
plt.plot(month_grid, predicted_age, color='blue', linewidth=3, label='Linear Regression')
plt.title('Month vs. Age')
plt.xlabel('Month')
plt.ylabel('Age')
plt.legend()
plt.show()
```
